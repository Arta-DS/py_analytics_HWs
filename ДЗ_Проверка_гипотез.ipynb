<div class="alert alert-info">
<b>Задание № 3</b>

Специалист по кадрам, работающий в технологической компании, интересуется сверхурочным временем разных команд. Чтобы выяснить, есть ли разница между сверхурочной работой команды разработчиков программного обеспечения и группы тестирования, она случайным образом выбрала 17 сотрудников в каждой из двух команд и записала их среднее сверхурочное время за неделю в пересчете на час. Данные ниже.

В соответствии с этой информацией проведите проверку гипотезы, чтобы проверить, есть ли разница между переутомлением двух команд, используя уровень значимости 0,05. Прежде чем приступать к проверке гипотез, проверьте предположение нормальности.
</div>

```python
from scipy import stats

test_team = [6.2, 7.1, 1.5, 2, 3, 2, 1.5, 6.1, 2.4, 2.3, 12.4, 1.8, 5.3, 3.1, 9.4, 2.3, 4.1]
developer_team = [2.3, 2.1, 1.4, 2.0, 8.7, 2.2, 3.1, 4.2, 3.6, 2.5, 3.1, 6.2, 12.1, 3.9, 2.2, 1.2 ,3.4]

# Проверка нормальности для команды тестировщиков
w, p_value_test_team = stats.shapiro(test_team)
print("Команда тестировщиков: w =", w, ", p-значение =", p_value_test_team)

# Проверка нормальности для команды разработчиков
w, p_value_dev_team = stats.shapiro(developer_team)
print("Команда разработчиков: w =", w, ", p-значение =", p_value_dev_team)
```

1. Формулировка гипотез:
- Нулевая гипотеза (H0): Среднее сверхурочное время команды тестировщиков равно среднему времени команды разработчиков.
- Альтернативная гипотеза (H1): Среднее сверхурочное время команды тестировщиков не равно среднему времени команды разработчиков.

2. Уровень значимости:
- Устанавливаем уровень значимости 0.05.

3. Проверка нормальности:
- Для команды тестировщиков:
  - Статистика W = 0.825, p-значение = 0.0046 (p < 0.05, нулевая гипотеза о нормальности отвергнута).
  
- Для команды разработчиков:
  - Статистика W = 0.753, p-значение = 0.0005 (p < 0.05, нулевая гипотеза о нормальности отвергнута).

Вывод: Оба набора данных не соответствуют нормальному распределению.

```python
import numpy as np

test_mean = np.mean(test_team)
test_std = np.std(test_team)

developer_mean = np.mean(developer_team)
developer_std = np.std(developer_team)

developer_mean, developer_std
```

```python
t_stat = (test_mean - developer_mean) / np.sqrt((test_std **2 / len(test_team)) + (developer_std **2 / len(developer_team)))
print(t_stat)
```

U-критерий Манна-Уитни (иногда называемый критерием суммы рангов Уилкоксона) используется для сравнения различий между двумя независимыми выборками, когда распределение выборки не является нормальным, а размеры выборки малы (n < 30).

```python
# Тест Манна-Уитни
u_statistic, p_value = stats.mannwhitneyu(test_team, developer_team, alternative='two-sided')
print("U-статистика =", u_statistic, ", p-значение =", p_value)
```

p-значение > 0.05, значит принимаем нулевую гипотезу

На основе данных теста Манна-Уитни нет статистически значимой разницы в среднем сверхурочном времени между командами тестировщиков и разработчиков. Оба команды имеют схожие показатели сверхурочной работы за неделю.

Педиатр хочет увидеть влияние потребления смеси на среднемесячную прибавку в весе у новорожденных. По этой причине она собрала данные из трех разных групп. Первая группа – дети исключительно грудного вскармливания, вторая группа – дети, которых кормят только смесью, и последняя группа – это дети, находящиеся на искусственном вскармливании и на грудном вскармливании. Эти данные приведены ниже.

В соответствии с этой информацией проведите проверку гипотезы, чтобы проверить, есть ли разница между среднемесячным приростом этих трех групп, используя уровень значимости 0,05. Если есть значительная разница, выполните дальнейший анализ, чтобы найти причину разницы.
</div>

```python
import numpy as np
from scipy.stats import ttest_ind

only_breast = [794.1, 716.9, 993. , 724.7, 760.9, 908.2, 659.3 , 690.8, 768.7, 717.3 , 630.7, 729.5, 714.1, 810.3, 583.5, 679.9, 865.1];
only_formula = [898.8, 881.2, 940.2, 966.2, 957.5, 1061.7, 1046.2, 980.4, 895.6, 919.7, 1074.1, 952.5, 796.3, 859.6, 871.1 , 1047.5, 919.1 , 1160.5, 996.9];
both = [976.4, 656.4, 861.2, 706.8, 718.5, 717.1, 759.8, 894.6, 867.6, 805.6, 765.4, 800.3, 789.9, 875.3, 740. , 799.4, 790.3, 795.2 , 823.6, 818.7, 926.8, 791.7, 948.3];

# Проверка гипотезы о равенстве средних значений для первой и второй группы
ttest_result = ttest_ind(only_breast, only_formula, equal_var=False)
print("Разница между средними значениями для групп 'только грудное вскармливание' и 'только смесь':")
print(f"t-статистика: {ttest_result.statistic}")
print(f"p-значение: {ttest_result.pvalue}")

# Проверка гипотезы о равенстве средних значений для второй и третьей группы
ttest_result = ttest_ind(only_formula, both, equal_var=False)
print("\nРазница между средними значениями для групп 'только смесь' и 'искусственное вскармливание + грудное вскармливание':")
print(f"t-статистика: {ttest_result.statistic}")
print(f"p-значение: {ttest_result.pvalue}")

# Проверка гипотезы о равенстве средних значений для первой и третьей группы
ttest_result = ttest_ind(only_breast, both, equal_var=False)
print("\nРазница между средними значениями для групп 'только грудное вскармливание' и 'искусственное вскармливание + грудное вскармливание':")
print(f"t-статистика: {ttest_result.statistic}")
print(f"p-значение: {ttest_result.pvalue}")
```

На основе полученных результатов можно сделать следующие выводы:

1. Есть статистически значимая разница между средним приростом веса у детей, находящихся на исключительно грудном вскармливании, и у детей, которых кормят только смесью. p-значение меньше 0.05, что говорит о том, что вероятность случайного возникновения такой разницы очень мала.

2. Также есть статистически значимая разница между приростом веса детей, которых кормят только смесью, и детей, находящихся на искусственном вскармливании и грудном вскармливании одновременно.

3. Однако разница между приростом веса детей на исключительно грудном вскармливании и детей на искусственном вскармливании и грудном вскармливании не является статистически значимой при уровне значимости 0.05. p-значение больше 0.05, поэтому нельзя исключить возможность того, что эта разница возникла случайно.
